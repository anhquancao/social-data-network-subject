# Comparing karate to Random Networks

## 2.1 From the average degree of karate and its number of nodes, compute the corresponding values of p in a random network having the same characteristics.

```python
degrees = [G.degree(n) for n in nx.nodes(G)]
num_nodes = len(degrees)
avg_degree = sum(degrees) / num_nodes
p = avg_degree / (num_nodes - 1)
```

```python
print(p)
```

```python
avg_degree
```

```python
RG = nx.gnp_random_graph(num_nodes, p)
```

```python
nx.draw(RG, with_labels=True)
```

```python
distances = nx.shortest_path_length(RG)
total = 0
count = 0
for node, distance in distances:
    values = distance.values()
    total += sum(values)
    count += len(values)
avg_dist = total / count
```

```python
# average distance
avg_dist
```

## Counting Triangles in Graph

```python
def triangles(G):
    count = 0
    A = nx.adjacency_matrix(G).todense()
    A3 = A @ A @ A
    res = dict()
    trace = A3.diagonal().T / 2
    for i in range(len(trace)):
        res[str(i)] = trace.item((i, 0))
    return res
```

```python
triangles(G)
```

```python
nx.triangles(G)
```

```python
def total_triangles_in_graph(G):
    count = 0
    A = nx.adjacency_matrix(G).todense()
    A3 = A @ A @ A
    trace = A3.diagonal()
    return np.sum(trace) / 6
    
```

```python
total_triangles_in_graph(G)
```

```python
# second way
sum(triangles(G).values()) / 3
```

## Compute the expected number of triangles in a random graph having the same average degree. How does it compare to the one found in reality?

```python
# There are 3Cn ways of select 3 nodes
num_selects = comb(num_nodes, 3)

# Prob to have 3 nodes form a triangle
prob = p ** 3

# Expected number of triangle in random graph
prob * num_selects # much smaller than reality 
```

# Community Detection

```python
def modularity(G, L):
    total = 0
    nodes = list(G.nodes())
    A = nx.adjacency_matrix(G).todense()
    for i in range(A.shape[0]):
        for j in range(A.shape[1]):
            pij = G.degree(nodes[i]) * G.degree(nodes[j]) / (2 * L)
            total += A.item((i, j)) - pij
    return total / (2 * L)
```

```python
def detect_communities(G):
    max_components = []
    max_M = -math.inf
    nodes = G.nodes()
    components = []
    L = G.number_of_edges()
    while (len(components) != len(nodes)):
        ebcs = nx.edge_betweenness_centrality(G)
        max_e = max(ebcs, key=ebcs.get)
        G.remove_edge(max_e[0], max_e[1])

        M = 0        
        for SG in components:
            M += modularity(SG, L)
            
        components = list(nx.connected_component_subgraphs(G))
#         print(M, len(components))

        if (M > max_M):
            max_M = M
            max_components = components
    return max_M, max_components
    
```

```python
G = nx.read_edgelist("karate.txt")
max_M, max_components = detect_communities(G)
```
